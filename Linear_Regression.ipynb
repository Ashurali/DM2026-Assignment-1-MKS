```python
# Sample Code for generating datasets
seed=42
np.random.seed(seed)
size = 500
dim = 1
noise_bound = 0.2
X,y,W = generate_linear_dataset(size,dim,noise_bound=noise_bound,is_reg=True)
plt.scatter(X[:,1],y)
```

```python
import os
from model.linear_model import LinearModel

datasets = ['linear_data_A', 'linear_data_B', 'linear_data_C', 'linear_data_D']
learning_rates = [0.1, 0.01, 0.001]
os.makedirs('Chart', exist_ok=True)

seed = 42  # do not change the seed

for dataset_name in datasets:
    print(f"\n{'='*60}")
    print(f"Dataset: {dataset_name}")
    print(f"{'='*60}")

    # Loading data
    loaded_data = np.load(f'data/linear_data/{dataset_name}.npz')
    X = loaded_data['X_data']
    y = loaded_data['y_data']
    W = loaded_data['W_data']
    print(f"X size : {X.shape}, y size : {y.shape}")

    size = X.shape[0]
    dim = X.shape[1] - 1

    # Split dataset to train dataset and validation dataset
    X_train, y_train, X_val, y_val = get_train_val(X, y)

    for lr in learning_rates:
        print(f"\n--- {dataset_name} | Learning Rate: {lr} | n_iteration: 500 ---")
        np.random.seed(seed)

        # Setting loss function and gradient
        loss_fn = MSE
        grad_fn = MSE_grad

        # Setting model and starting training
        model = LinearModel(dim, True, loss_fn, grad_fn)
        model.fit(X_train, y_train, lr=lr, n_iteration=500, val_ratio=0.2)
        plt.close('all')  # close the untitled inline loss plot from model.fit

        # Re-plot loss curve with title and save to file
        fig, ax = plt.subplots()
        ax.set_title(f'{dataset_name} | lr={lr} | Loss Curve')
        ax.set_xlabel('Iteration')
        ax.set_ylabel('Loss')
        ax.plot(range(len(model.train_losses)), model.train_losses, 'o-', color='r', label='Training loss', markersize=1)
        ax.plot(range(len(model.val_losses)), model.val_losses, 'o-', color='g', label='Validation loss', markersize=1)
        ax.legend(loc='best')
        fig.savefig(f'Chart/{dataset_name}_lr{lr}_loss_curve.png', dpi=150, bbox_inches='tight')
        plt.show()

        # Get the Prediction
        y_val_pred = model.predict(X_val)

        # Evaluation
        # Please complete the evaluate_linear_regression in the model/metrics.py by the sklearn package.
        evaluate_linear_regression(y_true=y_val, y_pred=y_val_pred, title=f'Linear Regression Evaluation | {dataset_name} | lr={lr}')

        # Plot
        fig, ax = plt.subplots()
        ax.set_title(f'{dataset_name} | lr={lr} | Prediction')
        ax.scatter(X_val[:, 1], y_val)
        ax.plot(X_val[:, 1], y_val_pred)
        fig.savefig(f'Chart/{dataset_name}_lr{lr}_prediction.png', dpi=150, bbox_inches='tight')
        plt.show()

        print()
```
